```python
import pandas as pd
import altair as alt
from altair import datum
```

```python
data = pd.read_csv('dataViz_formatted2.csv')
data.rename(columns={"Coverage subscore": "Coverage_SC", "Openness subscore": "Openness_SC", "Data categories": "Data_categories"}, inplace=True)
data
```

```python
data['Deviation'] = data['Coverage_SC'] - data['Openness_SC']
data['Deviation_norm'] = data['Deviation'] / abs(data['Deviation'])
data['Country'].unique().tolist()
```

```python
# Scatter Plot
points = alt.Chart(data).mark_point(filled=True).encode(
    x='Openness_SC:Q',
    y='Coverage_SC:Q',
    color=alt.Color('Deviation_norm:N', scale=alt.Scale(range=['red', 'blue'])),
    tooltip=['Country', 'Overall score', 'Coverage_SC', 'Openness_SC']
).transform_filter(
    (datum.Data_categories == 'All Categories')
)
points
```

```python
alt.Chart(data).mark_bar().encode(
    x="Country:N",
    y="Deviation:Q",
    color=alt.condition(
        alt.datum.Deviation > 0,
        alt.value("steelblue"),  # The positive color
        alt.value("red")  # The negative color
    ),
    tooltip=['Country', 'Overall score', 'Coverage_SC', 'Openness_SC']
).transform_filter(
    (datum.Data_categories == 'All Categories')
)
```

```python
countries = data['Country'].unique().tolist()
input_dropdown = alt.binding_select(options=countries, name='Country')
selection = alt.selection_single(fields=['Country'], bind=input_dropdown)
alt.Chart(data).mark_point(
    filled=True
).encode(
    x='Coverage_SC:Q',
    tooltip=['Country', 'Data_categories', 'Coverage_SC', 'Openness_SC']
).add_selection(
    selection
).transform_filter(
    selection
).properties(
    width=800
)
```
