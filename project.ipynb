```python
import pandas as pd
import altair as alt
from altair import datum
from vega_datasets import data
```

```python
odin_data = pd.read_csv('odin_data.csv')
odin_data.rename(columns={"Coverage subscore": "Coverage_SC", "Openness subscore": "Openness_SC", "Data categories": "Data_categories"}, inplace=True)
code_lookup = pd.read_csv("country_lookup.csv")
countries = alt.topo_feature(data.world_110m.url, 'countries')
odin_data
```

```python
code_lookup
```

```python
map_data = pd.DataFrame({
    'Country Code':odin_data[odin_data['Data_categories'].str.contains('All Categories')] ['Country Code'],
    'Overall':odin_data[odin_data['Data_categories'].str.contains('All Categories')] ['Overall score'],
    'Coverage':odin_data[odin_data['Data_categories'].str.contains('All Categories')] ['Coverage_SC'],
    'Openness':odin_data[odin_data['Data_categories'].str.contains('All Categories')] ['Openness_SC'],
})
merged_lookup_data = map_data.merge(code_lookup, left_on='Country Code', right_on='Alpha-3 code',how="outer").rename(columns={'Numeric code': 'id'})
merged_lookup_data = merged_lookup_data.drop(columns=['Alpha-2 code','Alpha-3 code', 'Country Code'])
merged_lookup_data
```

```python
odin_data['Deviation'] = odin_data['Coverage_SC'] - odin_data['Openness_SC']
odin_data['Deviation_norm'] = odin_data['Deviation'] / abs(odin_data['Deviation'])
#odin_data['Country'].unique().tolist()
odin_data
```

```python
odinmap = alt.Chart(countries).mark_geoshape(
    stroke='white',
).encode(
    color=alt.Color('Overall:O', scale=alt.Scale(range=['#D3D3D3', '#dbefdc', '#a6d7a8', '#5eb762', '#3d8c40', '#2e6930']), bin=True, legend=alt.Legend(title="ODIN Score:", orient="bottom-left")),
    #tooltip= alt.condition("!isValid(datum.Overall)", alt.Tooltip('Country:N'),alt.Tooltip(shorthand=['Country:N','Overall:Q'])),
    tooltip=['Country:N','Overall:Q'],
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(data=merged_lookup_data, key='id', fields=['Overall', 'Country','Openness','Coverage'])
).project(
    type='mercator',
    scale=570,
    center=[10,56],
    #center=[14,35] #MissingMalta
).properties(
#    width=650,
#    height=650
    width=650,
    height= 650
)
odinmap
```

```python
foreground = alt.Chart(countries).mark_geoshape().encode(
    color=alt.Color('Overall:O', scale=alt.Scale(range=['#D3D3D3', '#dbefdc', '#a6d7a8', '#5eb762', '#3d8c40', '#2e6930']), bin=True),
    tooltip=alt.Tooltip(shorthand=['Country:N','Overall:Q','Openness:Q','Coverage:Q']),
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(data=merged_lookup_data, key='id', fields=['Overall', 'Country','Openness','Coverage'])
).project(
    type='mercator',
    scale=250,
    center=[10,55],
    clipExtent= [[0,0], [400, 300]]
)

background = alt.Chart(countries).mark_geoshape(
    fill='lightgray',
    stroke='white'
).project(
    type='mercator',
    scale=250,
    center=[10,55],
    clipExtent= [[0,0], [400, 300]]
)

background + foreground
```

```python
# Scatter Plot
selection = alt.selection(type='single', encodings=['x'])
points = alt.Chart(odin_data).mark_point(filled=True).encode(
    x='Openness_SC:Q',
    y='Coverage_SC:Q',
    color=alt.condition( alt.datum.Deviation > 0, alt.value("#4682b4"),alt.value("#b91f13")),
    opacity = alt.condition(selection, alt.value(1.0), alt.value(0.2)),
    #color=alt.condition(selection, alt.Color('Deviation_norm:N', scale=alt.Scale(range=['red', 'blue'])), alt.ColorValue('gray')),
    tooltip=['Country', 'Overall score', 'Coverage_SC', 'Openness_SC'],
).transform_filter(
    (datum.Data_categories == 'All Categories')
).add_selection(
    selection
).properties(
    width=650,
    height=650
)
points
```

```python
bars = alt.Chart(odin_data).mark_bar().encode(
    x="Country:N",
    y="Deviation:Q",
    color=alt.Color('Deviation_norm:N', scale=alt.Scale(range=['#b91f13', '#4682b4']),legend=alt.Legend(title="Deviation")),
    #color=alt.condition(selection, alt.Color('Deviation_norm:N', scale=alt.Scale(range=['#b91f13', '#4682b4'])), alt.value("grey")),
    opacity = alt.condition(selection, alt.value(1.0), alt.value(0.2)),
    tooltip=['Country', 'Overall score', 'Coverage_SC', 'Openness_SC']
).transform_filter(
    (datum.Data_categories == 'All Categories')
).add_selection(
    selection
).properties(
    width=650,
    height=650
)
bars
```

```python
points | bars
```
